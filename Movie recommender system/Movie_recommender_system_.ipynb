```python
# This --> [{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]
# Convert to ["Action", "Adventure", "Fantasy", "Science Fiction", ]

import ast  ## module in python use to convert string object into list

def convert(obj):
  L = []
  for i in ast.literal_eval(obj):
    L.append(i["name"])
  return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
def convert3(obj):
  L = []
  counter = 0
  for i in ast.literal_eval(obj):
    if counter != 3:
      L.append(i["name"])
      counter += 1
    else:
      break

  return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
def fetch_director(obj):
  L = []
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
      L.append(i['name'])
      break
  return L
```

```python
movies['crew']= movies['crew'].apply(fetch_director)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres'] =  movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['keywords'] =  movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['cast'] =  movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['crew'] =  movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_movies_df = movies[['movie_id', 'title', 'tags']]
```

```python
new_movies_df.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []

  for i in text.split():
    y.append(ps.stem(i))

  return " ".join(y)
```

```python
new_movies_df['tags']= new_movies_df['tags'].apply(stem)
```

```python
new_movies_df['tags']= new_movies_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_movies_df['tags']= new_movies_df['tags'].apply(lambda x:x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new_movies_df['tags']).toarray()
```

```python
vector.shape
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
new_movies_df[new_movies_df['title'] == 'The Lego Movie'].index[0]
```

```python
def recommend(movie):
    index = new_movies_df[new_movies_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_movies_df.iloc[i[0]].title)
```

```python
recommend('John Carter')
```
